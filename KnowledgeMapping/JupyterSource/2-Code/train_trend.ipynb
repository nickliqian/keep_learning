```python
import matplotlib.pyplot as plt
from scipy.interpolate import spline
```

```python
fig = plt.figure(figsize=(10, 8))
colors = ["red", "blue", "green"]

plt.plot(df["step"], df["acc_char"], c="red", label="acc_char")
plt.plot(df["step"], df["acc_image"], c="blue", label="acc_image")
plt.plot(df["step"], df["cost_"], c="green", label="cost_")

plt.legend(loc="upper left")
plt.ylabel("acc_char, acc_image, loss")
plt.xlabel("step")
plt.title("Train set acc trend")
```

```python
fig = plt.figure(figsize=(10, 8))
colors = ["red", "blue", "green"]

xnew = np.linspace(df["step"].min(),df["step"].max(),300)

power_smooth1 = spline(df["step"], df["acc_char"], xnew)
power_smooth2 = spline(df["step"], df["acc_image"], xnew)
power_smooth3 = spline(df["step"], df["cost_"], xnew)

plt.plot(power_smooth1, c="red", label="acc_char")
plt.plot(power_smooth2, c="blue", label="acc_image")
plt.plot(power_smooth3, c="green", label="cost_")

plt.legend(loc="upper left")
plt.ylabel("acc_char, acc_image, loss")
plt.xlabel("step")
plt.title("Train set acc trend")

```
