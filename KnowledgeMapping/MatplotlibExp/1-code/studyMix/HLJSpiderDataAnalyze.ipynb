```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/nick/Desktop/DA/speedFile/count.csv", index_col=None, names=["total"], 
                 parse_dates=True, dtype={"total": np.int64}, encoding="utf-8")
print(df.shape)
print(df.dtypes)
df = df.loc["2018-01-02 00:00:00":"2018-01-12 00:00:00"]
```

```python
# 按照原始数据显示 显示总数
df.plot()
plt.show()
```

```python
# 降低采样频率 以一小时为单位显示总量变化
a = df.resample('1H').mean()
a.plot()
plt.show()
```

```python
# 以一小时为单位显示采集速度
increment = df['total'].diff()
dif = pd.DataFrame({"crawl speed": increment}, index=df.index)
b = dif.resample('1H').sum()
b.plot()
plt.show()
```

```python
# 以两小时为单位显示采集速度
c = dif.resample('2H').sum()
c.plot()
plt.show()
```

```python
# 显示每天抓取的量
dif.plot(kind="bar")
plt.show()
```
