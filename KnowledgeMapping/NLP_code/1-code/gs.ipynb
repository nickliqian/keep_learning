```python
import pandas as pd
import numpy as np
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
df = pd.read_csv(r"D:\dataset\NewsFT_Ret_500.csv")
df.head()
```

```python
df.loc[df["FAANG_Ret_NextTD"] <= 0,'FAANG_Ret_NextTD'] = 0
df.loc[df["FAANG_Ret_NextTD"] > 0,'FAANG_Ret_NextTD'] = 1
df.head()
```

```python
columns = ["tok_content","label"]
df_new = pd.DataFrame(columns=columns)
for i in range(df.shape[0]):
    df_new.loc[i,"label"] = df.loc[i,"FAANG_Ret_NextTD"]
    df_new.loc[i,"tok_content"]=word_tokenize(df.loc[i,"news"])
df_new.head()
```

```python
word_tokenize(df.loc[10,"news"])
```

```python
import gensim
allTokens = []
for tok_content in df_new.tok_content:
    allTokens.append(tok_content)
```

```python
len(allTokens)
```

```python
word2Vect_model = gensim.models.Word2Vec(allTokens, min_count=1)
word2Vect_model.most_similar("bright")
```

```python
word2idx = dict()
idx2word = list()
for tok_content in df_new.tok_content:
    for word in tok_content:
        if word not in word2idx:
            idx2word.append(word)
            word2idx[word] = len(idx2word) - 1

df_new.head()
```

```python
idx2word
```

```python
for i in range(df_new.shape[0]):
    data = [word2idx[token] for token in df_new.loc[i,'tok_content']]
    df_new.loc[i,"tok_content"] = data
    
df_new.head()
```

```python
import jieba
```

```python
raw_sentences = ["但由于中文没有像英文那么自带天然的分词","所以我们第一步采用分词"]
```

```python
sentences = []
for s in raw_sentences:
    tmp = []
    for item in jieba.cut(s):
        tmp.append(item)
    sentences.append(tmp)

sentences
```

```python
model = gensim.models.Word2Vec(sentences, min_count=1)
```
